```python
import matplotlib.pyplot as plt 
import numpy as np
import math
%matplotlib inline
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
ax.set_title("Onda")
ax.set_xlabel("Angulos")
ax.set_ylabel("Seno")
```
